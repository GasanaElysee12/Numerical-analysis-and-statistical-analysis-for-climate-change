Solving $y^\prime=-\lambda y,\quad u(t=0)=1.0,\ \lambda=0.1 s^{-1}$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

def PlotterForward(uu,tt,color=None):
    plt.plot(tt,uu,color=color,label='approx')
#     plt.plot(tt,np.cos(omega*tt),'r--',label='True')
```

Q_1. Let's solve this differential equation $y^\prime=-\lambda y .$

By using forward finite difference method.

$y_{n+1}=(1-\lambda \Delta t)y_n, \quad n=0,1,2,3,\cdots$

```python
def forwardMethod(t_0, y_0,dt,lamb, t_N):
    t=np.arange(t_0,t_N+dt,dt) #initialize time
    y_approx=np.zeros(int(len(t))) #initialize y
    y_approx[0]=y_0 #Assign y_0 at first index in y
    
    #iterate to get new numerical value of y
    for n in range(0,len(t)-1):
        y_approx[n+1]=(1-lamb*dt)*y_approx[n]
        
    y_true=y_0*np.exp(-lamb*t)
    #PlotterForward(y,t)
    #print('t=',len(t),'y_true=',len(y_true),'y_approx=',len(y_approx))
    
    PlotterForward(y_approx-y_true,t,color='r')
    #PlotterForward(g,t,color='g')
```

```python

dt=[0.1,15,25]
for idx in range(len(dt)):
    plt.figure(figsize=(12,15))
    plt.subplot(3,1,idx+1)
    plt.title(r'$y=e^{-\lambda t}$ approximation error with $\Delta t=$'+str(dt[idx]))
    forwardMethod(t_0=0,y_0=1.0,dt=dt[idx],lamb=0.1,t_N=200)
    plt.xlabel('Time (s)')
    plt.ylabel(r'Error, $y-y_{approximate}$')
    plt.grid(True)
plt.show()
```

Q_2.Let's solve the differential equation using a backward difference method.
 $y^\prime=-\lambda y$ while its approximation will be $y_{n}=\big (\frac{1}{1+\lambda \Delta t}\big )y_{n-1},\quad n=1,2,3,\cdots$

```python
def BackwardMethod(t_0, y_0,dt,lamb, t_N):
    t=np.arange(t_0,t_N+dt,dt) #initialize time
    y_approx=np.zeros(int(len(t))) #initialize y
    y_approx[0]=y_0 #Assign y_0 at first index in y
    
    #iterate to get new numerical value of y
    for n in range(0,len(t)-1):
        y_approx[n+1]=(1/(1+lamb*dt))*y_approx[n]
        
    y_true=y_0*np.exp(-lamb*t)# true value of y
    
    #PlotterForward(y_true,t)
    print('t=',len(t),'y_true=',len(y_true),'y_approx=',len(y_approx))
    PlotterForward(y_approx-y_true,t,color='r')
    #PlotterForward(g,t,color='g')
```

```python
dt=[0.1,15,25]
plt.style.use('ggplot')
for idx in range(len(dt)):
    plt.figure(figsize=(12,15))
    plt.subplot(3,1,idx+1)
    plt.title(r'$y=e^{-\lambda t}$ approximation error with $\Delta t=$'+str(dt[idx]))
    BackwardMethod(t_0=0,y_0=1.0,dt=dt[idx],lamb=0.1,t_N=200)
    plt.xlabel('Time (s)')
    plt.ylabel(r'Error, $y-y_{approximate}$')
    plt.grid(True)
plt.show()
```
